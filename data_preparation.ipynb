```python
import pandas as pd
all_vars_clean = pd.melt(
    all_vars,
    id_vars=['Country code', 'Country name', 'Series code', 'Series name', 'SCALE', 'Decimals'],
    var_name='year',
    value_name='value'
)

# Step 3: Convert 'year' column to numeric
all_vars_clean['year'] = pd.to_numeric(all_vars_clean['year'], errors='coerce')

# Step 4: Count missing values for each year
missing_values_per_year = all_vars_clean.groupby('year')['value'].apply(lambda x: x.isnull().sum())

# Step 5: Print missing values per year
print("Missing values by year:")
for year, count in missing_values_per_year.items():
    print(int(year), ":", count)

```

```python
# Print number of missing values and rows before filtering
print("number of missing values in the whole dataset before filtering the years:")
print(all_vars_clean.isnull().sum().sum())

print("number of rows before filtering the years:")
print(all_vars_clean.shape[0])

# Filter only rows for years between 1991 and 2008 (inclusive)
all_vars_clean = all_vars_clean[(all_vars_clean['year'] >= 1991) & (all_vars_clean['year'] <= 2008)]

# Print number of missing values and rows after filtering
print("number of missing values in the whole dataset after filtering the years:")
print(all_vars_clean.isnull().sum().sum())

print("number of rows after filtering the years:")
print(all_vars_clean.shape[0])
```

```python
# Group by 'Country name' and count missing 'value' entries
missing_values_per_country = all_vars_clean.groupby('Country name')['value'].apply(lambda x: x.isnull().sum())

# Sort the result (optional)
missing_values_per_country = missing_values_per_country.sort_values()

# Print missing values by country
print("Missing values by country:")
for country, count in missing_values_per_country.items():
    print(country, ":", count)
```

```python
# Step 1: Count missing values per country
missing_by_country = all_vars_clean.groupby('Country name')['value'].apply(lambda x: x.isnull().sum())

# Step 2: Create list of countries with zero missing values
countries_filter = []
for key, val in missing_by_country.items():
    if val == 0:
        countries_filter.append(key)

# Step 3: Filter the dataset for those countries
all_vars_clean = all_vars_clean[all_vars_clean['Country name'].isin(countries_filter)]

# Step 4: Print stats before/after filtering
print("number of missing values in the whole dataset after filtering the countries:")
print(all_vars_clean.isnull().sum().sum())

print("number of rows after filtering the countries:")
print(all_vars_clean.shape[0])
```

```python
# Count missing values in each column
missing_values_per_column = all_vars_clean.isnull().sum()

# Print results
print("Remaining missing values for each column:")
print(missing_values_per_column)
```
